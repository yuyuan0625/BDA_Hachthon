### 讀入3個csv

```python
import pandas as pd
# Read the Excel file into a Pandas dataframe
df1 = pd.read_csv("201811_ubike_ticket_001.csv")
df1 = df1.set_axis(["交易序號","卡號","扣款時間","扣款金額","借車時間","借車站代號","借車場站","借車車柱","自行車編號","還車時間","還車站代號","還車場站","還車車柱","租用","手機序號","卡片種類","費率別","業者代碼","根據檔案是IPASS或","整批資料更新時間","存入DB時間","該筆資料更新時間","該筆資料更新時間"], axis=1, inplace=False)
df2 = pd.read_csv("201811_ubike_ticket_002.csv")
df2 = df2.set_axis(["交易序號","卡號","扣款時間","扣款金額","借車時間","借車站代號","借車場站","借車車柱","自行車編號","還車時間","還車站代號","還車場站","還車車柱","租用","手機序號","卡片種類","費率別","業者代碼","根據檔案是IPASS或","整批資料更新時間","存入DB時間","該筆資料更新時間","該筆資料更新時間"], axis=1, inplace=False)
df3 = pd.read_csv("201811_ubike_ticket_003.csv")
df3 = df3.set_axis(["交易序號","卡號","扣款時間","扣款金額","借車時間","借車站代號","借車場站","借車車柱","自行車編號","還車時間","還車站代號","還車場站","還車車柱","租用","手機序號","卡片種類","費率別","業者代碼","根據檔案是IPASS或","整批資料更新時間","存入DB時間","該筆資料更新時間","該筆資料更新時間"], axis=1, inplace=False)
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

### 合併

```python
merge = pd.concat([df1,df2,df3],axis=0,ignore_index = True )
print(merge.shape)
merge
```

### 借車、還車資料

```python
data1 = merge[["借車時間","借車場站"]]
df_out = data1.copy()
# Convert the date column to a Pandas datetime format
df_out['借車時間'] = pd.to_datetime(df_out['借車時間'])
df_out = df_out[df_out['借車場站'] == '捷運公館站(2號出口)']
df_out
```

```python
#確保資料有在時間範圍內
df_out = df_out[df_out['借車時間'] >= '2018-11-01 00:00:00']
df_out = df_out[df_out['借車時間'] <= '2018-11-30 23:59:59']
df_out = df_out.sort_values(by='借車時間')
df_out['value'] = 1
df_out
```

```python
out_resample = df_out.set_index('借車時間').resample('H')['value'].sum()
df_out_resample = pd.DataFrame(out_resample)
df_out_resample = df_out_resample.set_axis(["借車人次"], axis=1, inplace=False)
# df_out_resample.reset_index()
df_out_resample
```

```python
data2 = merge[["還車時間","還車場站"]]
df_in = data2.copy()
# Convert the date column to a Pandas datetime format
df_in['還車時間'] = pd.to_datetime(df_in['還車時間'])
df_in = df_in[df_in['還車場站'] == '捷運公館站(2號出口)']
df_in
```

```python
df_in = df_in[df_in['還車時間'] >= '2018-11-01 00:00:00']
df_in = df_in[df_in['還車時間'] <= '2018-11-30 23:59:59']
df_in = df_in.sort_values(by='還車時間')
df_in['value'] = 1
df_in
```

```python
in_resample = df_in.set_index('還車時間').resample('H')['value'].sum()
df_in_resample = pd.DataFrame(in_resample)
df_in_resample = df_in_resample.set_axis(["還車人次"], axis=1, inplace=False)
# df_in_resample.reset_index(drop=True)
df_in_resample
```

```python
start_date = "2018-11-01 00:00:00"
end_date = "2018-11-30 23:00:00"
date_range = pd.date_range(start=start_date, end=end_date, freq="H")
res = pd.DataFrame({'時間1': date_range})
res["時間"] =date_range
res = res.set_index('時間1')
res
```

```python
res = pd.concat([res,df_out_resample,df_in_resample], axis=1, ignore_index = True )
res = res.set_axis(["時間", "借車人數","還車人數"], axis=1, inplace=False)
res['該小時車輛變化'] = res['還車人數'] - res['借車人數']
res
```

```python
# save the DataFrame to a CSV file
res.to_csv('公館201811_車輛變化數.csv', index=False)
```
